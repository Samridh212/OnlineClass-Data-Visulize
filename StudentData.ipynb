Welcome To the data Visulization of Online class Adaptivity

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('students_adaptability_level_online_education.csv')
```

```python
df.head(6)
```

```python
plt.figure(figsize=(25,15))
plt.subplot(2,2,1)
un , count = np.unique(df[df['Gender']=='Boy']['Adaptivity Level'].values,return_counts=True)
plt.bar(un,count,color='blue')
plt.title("Adaptivity Level for Males")
plt.xlabel("Adaptivity Level")
plt.ylabel("Count")

plt.subplot(2,2,2)
un , count = np.unique(df[df['Gender']=='Girl']['Adaptivity Level'].values,return_counts=True)
plt.bar(un,count,color='pink')
plt.title("Adaptivity Level for Females")
plt.xlabel("Adaptivity Level")
plt.ylabel("Count")

plt.tight_layout()
```

```python
plt.figure(figsize=(15,10))
graphs = 1
for i in np.unique(df['Age'].values):
    plt.subplot(2,3,graphs)
    un, count = np.unique(df[df.Age == i]["Adaptivity Level"].values, return_counts=True)
    count, un = zip(*sorted(zip(count, un), reverse=True))
    plt.bar(un, count, color = ['yellow','green','red'])
    plt.title("Adaptivity Level for Age group " + i)
    plt.xlabel("Age group")
    plt.ylabel("Count")
    graphs = graphs + 1
plt.tight_layout()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
un , count = np.unique(df[df['IT Student']=='Yes']['Adaptivity Level'].values,return_counts=True)
plt.bar(un,count,color = ['yellow','green','red'])
plt.title("IT Student")
plt.xlabel("Adaptivity Level")
plt.ylabel("Count")

plt.subplot(2,2,2)
un , count = np.unique(df[df['IT Student']=='No']['Adaptivity Level'].values,return_counts=True)
plt.bar(un,count,color = ['yellow','green','red'])
plt.title("Non IT Student")
plt.xlabel("Adaptivity Level")
plt.ylabel("Count")

plt.tight_layout()
```
